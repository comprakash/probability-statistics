```python
from itertools import product
```

```python
dice_four_rolls_combinations = list(product(range(1, 7), repeat=4))
```

```python
dice_four_rolls_combinations
```

```python
len(dice_four_rolls_combinations)
```

```python
atleast_one_six = [c for c in dice_four_rolls_combinations if 6 in c]
```

```python
atleast_one_six
```

```python
len(atleast_one_six)
```

```python
prob_atleast_one_six = len(atleast_one_six) / len(dice_four_rolls_combinations)
```

```python
prob_atleast_one_six
```
